```python
from dicompylercore import dicomparser, dvh, dvhcalc, dose
from scipy.integrate import cumulative_trapezoid, trapezoid
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.pyplot as plt
from pydicom import dcmread
from pydicom.data import get_testdata_file
import scienceplots

plt.style.use("science")
dp = dicomparser.DicomParser("../CTData/RTSTruct.dcm")

# i.e. Get a dict of structure information
structures = dp.GetStructures()
```

```python
for n in range(50000, 250000, 50000):
    dose_grid= dose.DoseGrid(f"./no_protons/outputs/job{n}.dcm")
    for j in range(1,8):
        dose_grid.multiply(1/np.max(dose_grid.dose_grid))
        dose_grid.save_dcm(f"./no_protons/outputs/job{n}_normalised.dcm")
        calcdvh = dvhcalc.get_dvh("../CTData/RTSTruct.dcm", f"./no_protons/outputs/job{n}_normalised.dcm", j)
        print(trapezoid(calcdvh.relative_volume.counts, calcdvh.bincenters ))
        print(len(calcdvh.bincenters))
        plt.step(
            calcdvh.bincenters,
            calcdvh.relative_volume.counts,
            label=structures[j]["name"]
        )
        print(calcdvh.describe())
    plt.ylim(-1,102)
    plt.legend()
    plt.show()
```

```python
auc = [[] for i in range(0,7)]

bragg_peaks = []
beam_widths = []

for i in range(20,420, 20):
    plt.figure(figsize=(8,6))
    for j in range(1,8):
        dose_grid= dose.DoseGrid(f"./outputs/job{i}.dcm")
        dose_grid.multiply(1/np.max(dose_grid.dose_grid))
        dose_grid.save_dcm(f"./outputs/job{i}_normalised.dcm")
        calcdvh = dvhcalc.get_dvh("../CTData/RTSTruct.dcm", f"./outputs/job{i}_normalised.dcm", j)
        auc[j - 1].append(trapezoid(calcdvh.relative_volume.counts, calcdvh.bincenters))
        arr = []
    
        plt.step(
            calcdvh.bincenters,
            calcdvh.relative_volume.counts,
            label=structures[j]["name"]
        )
        #print(calcdvh.describe())
    plt.ylim(-1,102)
    plt.legend()
    plt.ylabel(r"Volume / $\%$", fontsize=16)
    plt.xlabel("Dose / arb", fontsize=16)
    plt.savefig(f"./plots/mb_DVH_{i}.png", dpi=300)
    
    plt.show()
    
    arr = []

    for k in range(1,5):
        arr.append(dcmread(f"../CTData/CT{k}.dcm").pixel_array)
    arr = np.sum(arr, axis = 0)

    fig,ax = plt.subplots(1, figsize=(6,6))
    print(f"{i} MeV")
    ax.imshow(arr, cmap="gray", alpha=1)
    mesh = ax.pcolormesh([i for i in range(128)], [i for i in range(128)], np.flip(np.flip(np.sum(dose_grid.dose_grid, axis=2).T, axis=0), axis=0), alpha=0.5, vmin=0, vmax=1)
    cbar = fig.colorbar(mesh, fraction=0.046, pad=0.04)
    cbar.set_label("Dose", fontsize=16)
    ax.set_xlabel("X pixel", fontsize=16)
    ax.set_ylabel("Y pixel", fontsize=16)
    # Add axis on the side
    plt.savefig(f"./plots/dose_in_patient_{i}.png", dpi=300)
    # Add axis on the side
    plt.show()
    doses = np.sum(np.flip(np.sum(dose_grid.dose_grid, axis=2)).T, axis=1)

    max_pos = np.argmax(doses)
    bragg_peaks.append(max_pos)
    idxs = np.setdiff1d(np.arange(len(doses)), np.where(doses < doses[max_pos]/np.exp(2)))
    beam_widths.append((max_pos - idxs[0]))
```

```python
plt.figure(figsize=(8,6))
plt.errorbar(
    [i for i in range(100, 106, 1)],
    128 - np.array(fine_bragg_peaks),
    yerr=fine_beam_widths,
    c="black",
    ls="None",
    marker="o",
    label="Beam width"
)
plt.errorbar(
    [i for i in range(20, 420, 20)],
    128 - np.array(bragg_peaks),
    yerr=beam_widths,
    c="black",
    ls="None",
    marker="o"
)

plt.axhline(y = 64 + 46/4, c="red", ls="--")
plt.axhspan(ymin = 64 + 46/4 - 30/4, ymax=64 + 46/4 + 30/4, color="red", alpha=.5, label="Tumour")

plt.xlabel("Energy / MeV", fontsize=16)
plt.ylabel("Bragg peak position / pixels", fontsize=16)
plt.legend()
plt.xlim(15, 145)
plt.ylim(35, 100)
plt.savefig("./plots/bragg_peaks_energy.png", dpi=300)
```
